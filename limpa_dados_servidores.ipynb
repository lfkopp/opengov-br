## Cleaning data

```python
import pandas as pd
from unicodedata import normalize

def cpf_mask_remove(pseudo_cpf):
    pseudo_cpf = str(pseudo_cpf)
    c = str(pseudo_cpf.replace("*","").replace(".","").replace("-",""))
    if len(str(c)) > 6:
        x1 = 3 - (11 - len(str(c)))
        x2 = x1 + 6
        return c[x1:x2]
    else:
        return c

def remover_acentos(txt):
    return normalize('NFKD', txt).encode('ASCII', 'ignore').decode('ASCII')
```

```python
emp = ('cnpj,razao,uf').split(',')
cnpj = pd.read_pickle('cnpj.pickle')[emp]
cnpj.tail(15)
```

```python
soc = ('nome,quadro,cnpj').split(',')
socios = pd.read_pickle('socios.pickle')[soc]
socios.tail(15)
```

```python
cand = ('ano_eleicao,sigla_uf,descricao_cargo,numero_candidato,cpf_candidato,nome_candidato,nome_urna_candidato,cod_situacao_candidatura,'
        'numero_partido,sigla_partido,descricao_ocupacao,data_nascimento,idade_data_eleicao,descricao_sexo,'
        'descricao_cod_instrucao,descricao_estado_civil,descricao_nacionalidade,'
        'sigla_uf_nascimento,nome_municipio_nascimento,despesa_max_campanha,desc_sit_tot_turno').split(',')
candidatos_2016 = pd.read_pickle('candidatos_2016.pickle')[cand]
candidatos_2014 = pd.read_pickle('candidatos_2014.pickle')[cand]
candidatos_2012 = pd.read_pickle('candidatos_2012.pickle')[cand]
candidatos_2010 = pd.read_pickle('candidatos_2010.pickle')[cand]
#candidatos_2008 = pd.read_pickle('candidatos_2008.pickle')[cand]
#candidatos_2006 = pd.read_pickle('candidatos_2006.pickle')[cand]
#candidatos_2004 = pd.read_pickle('candidatos_2004.pickle')[cand]
#candidatos_2002 = pd.read_pickle('candidatos_2002.pickle')[cand]
#candidatos_2000 = pd.read_pickle('candidatos_2000.pickle')[cand]
```

```python
candidatos = pd.concat([candidatos_2016,candidatos_2014,candidatos_2012,candidatos_2010])
candidatos['nome_limpo'] = candidatos['nome_candidato'].apply(remover_acentos)
candidatos.tail(15)
```

```python
socio_empresa = pd.merge(socios,cnpj,how='inner',on='cnpj',suffixes=('_socio','_empresa'))
socio_empresa.tail(15)
```

```python
socio_candidato = pd.merge(candidatos,socio_empresa,how='inner',left_on='nome_limpo',right_on='nome',
                           suffixes=('_candidato','_empresa'))
socio_candidato.tail(15)
```

```python
servidores = pd.read_pickle('servidores.pickle')
servidores['cpf2'] = servidores['CPF'].apply(cpf_mask_remove)
servidores.tail(15)
```

```python
candidatos['cpf2'] = candidatos['cpf_candidato'].apply(cpf_mask_remove)
candidato_servidor = pd.merge(candidatos,servidores,how='inner',left_on='nome_limpo', right_on='NOME', suffixes=('_candidato','_servidor'))
candidato_servidor = candidato_servidor[candidato_servidor['cpf2_servidor'] == candidato_servidor['cpf2_candidato']]
#candidato_servidor.columns.values.tolist()
candidato_servidor
```

```python
remuneracao = pd.read_pickle('remuneracao.pickle')
remuneracao['cpf2'] = remuneracao['CPF'].apply(cpf_mask_remove)
#remuneracao.columns.values.tolist()
candidato_servidor_salario = pd.merge(candidato_servidor,remuneracao,how='left',left_on='nome_limpo',
                                      right_on='NOME', suffixes=('_servidor','_remuneracao'))

candidato_servidor_salario = candidato_servidor_salario[candidato_servidor_salario['cpf2'] == candidato_servidor_salario['cpf2_servidor']]
candidato_servidor_salario
```
