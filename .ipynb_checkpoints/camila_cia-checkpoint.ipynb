```python
import requests

months = ('January,February,March,April,May,June,July,August,September,October,November,December').split(',')
years = list(range(2001,2018))

for year in years:
    print('\n'+str(year), end= ' ')
    for m in range(12):
        month = months[m]
        if m <9: m = '0'+str(m+1)
        else:    m = str(m+1)
        try:
            url = 'https://www.cia.gov/library/publications/resources/world-leaders-1/pdfs/'+str(year)+'/'+month+str(year)+'ChiefsDirectory.pdf'
            r = requests.get(url)
            with open('./cia/'+str(year)+'-'+m+'.pdf', 'wb') as f:
                f.write(r.content)
                print('.',end='')
        except:
            print('o',end='')
```
